```python
d1 = {"Xiaomi Mi 11": 0, "iPhone 12":1, "Google Pixel 5":2, "OnePlus 9":3, "Samsung Galaxy S21":4}
df["Device Model"] = df["Device Model"].map(d1)
```

```python
df["Operating System"].value_counts()
```

```python
d2 = {"Android":0, "iOS":1}
df["Operating System"] = df["Operating System"].map(d2)
```

```python
d3 = {"Male":0, "Female":1}
df["Gender"] = df["Gender"].map(d3)
```

```python
df.head()
```

```python
x = df.drop(columns="Gender")
y = df["Gender"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
#Prediction

y_pred = lr.predict(x_test)
```

```python
# Evalution

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
y_pred1 = linear.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,y_pred1)
```

```python
mse
```

```python
rmse = np.sqrt(mse)
```
